We will now move on to explore the effect of the crossover rate. We will plot the best and average fitness values vs. the generation. We will repeat the expermentations for different seed values. 

```python
from genetic_algorithm import GeneticAlgorithm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We will define a function to automate the process for us

```python
def TestCrossover(crossovers, seeds):
    ga = GeneticAlgorithm(
        110,
        0.1,
        0.5,
        0.5,
        60
    )
    solutions = {}
    for i in range(len(seeds)):
        seed = seeds[i]
        for j in range(len(crossovers)):
            ga.crossover_rate = crossovers[j]
            x, solutions["best_fitness_values_" + str(crossovers[j]) + "_" + str(seed)], solutions["average_fitness_values_" + str(crossovers[j]) + "_" + str(seed)] = ga.evolve(seed)
    return solutions
```

```python
crossovers = [0.25, 0.5, 0.75, 1]
seeds = [42, 80]
solutions = TestCrossover(crossovers, seeds)
```

```python
print(solutions.keys())
```

```python
generations = np.arange(0, 60).reshape(-1, 1)
for i in solutions:
    fig, ax = plt.subplots()
    
    i_split = i.split("_")
    if "best" in i_split:
        plt.title(label=("Best Fitness Vs. Generation" + " Crossover=" + i_split[3] + " Seed=" + i_split[4]))
    else:
        plt.title(label=("Average Fitness Vs. Generation" + " Crossover=" + i_split[3] + " Seed=" + i_split[4]))
    plt.plot(generations, solutions[i])
    plt.show()
```

```python
def TestMutations(mutations, seeds):
    ga = GeneticAlgorithm(
        110,
        0.1,
        0.5,
        0.5,
        60
    )
    solutions = {}
    for i in range(len(seeds)):
        seed = seeds[i]
        for j in range(len(mutations)):
            ga.mutation_rate = mutations[j]
            x, solutions["best_fitness_values_" + str(mutations[j]) + "_" + str(seed)], solutions["average_fitness_values_" + str(mutations[j]) + "_" + str(seed)] = ga.evolve(seed)
    return solutions
mutations = [0.25, 0.5, 0.75, 1]
seeds = [42, 80]
solutions = TestMutations(crossovers, seeds)

generations = np.arange(0, 60).reshape(-1, 1)
for i in solutions:
    fig, ax = plt.subplots()
    
    i_split = i.split("_")
    if "best" in i_split:
        plt.title(label=("Best Fitness Vs. Generation" + " Mutation=" + i_split[3] + " Seed=" + i_split[4]))
    else:
        plt.title(label=("Average Fitness Vs. Generation" + " Mutation=" + i_split[3] + " Seed=" + i_split[4]))
    plt.plot(generations, solutions[i])
    plt.show()
```
